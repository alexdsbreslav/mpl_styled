```python
import quicklook
import numpy as np
import pandas as pd
```

```python
pd.Series(np.random.normal(0,0.5,10)).sort_values()
```

```python
chart_skeleton = quicklook.build_chart_skeleton(style = 'default', size = 'default',
title = 'Visualize Your Data With quicklook',
xlabel = 'Some Arbitrary Label',
ylabel = 'Another\nArbitrary Label',
x_min_max = (0,1), y_min_max = (0,1),
xtick_interval = 0.5, ytick_interval = 0.25,
horizontal_gridlines_on = False,
vertical_gridlines_on = False);

quicklook.add_scatter_to_chart(chart_skeleton,
x = np.random.uniform(0.5,1,20),
y = np.random.uniform(0,0.75,20),
color_name = 'orange',
color_brightness = 'default',
marker_shape = 'o',
opacity = 1,
label_for_legend = 'Orange Dots',
layer_order = 3)

quicklook.add_scatter_with_error_to_chart(chart_skeleton,
x = np.random.uniform(0,0.5,10),
y = np.random.uniform(0.25,1,10),
x_error = np.full(10,0.05),
y_error = np.full(10,0.05),
color_name = 'indigo',
color_brightness = 'default',
marker_shape = 'o',
label_for_legend = 'Blue Clouds',
layer_order = 2)

quicklook.add_line_to_chart(chart_skeleton,
x = np.linspace(0,1,100),
y = 1/(1+np.exp(-8*np.linspace(-1,1,100))),
color_name = 'gray',
color_brightness = 'dark',
linewidth = 3,
linestyle = '-.',
marker_shape = '',
opacity = 1,
label_for_legend = '',
layer_order = 1)
```

```python
chart_skeleton = quicklook.build_chart_skeleton(style = 'default', size = 'default',
title = '',
xlabel = '',
ylabel = '',
x_min_max = (-1.1,1.1), y_min_max = (-1.9,1.9),
xtick_interval = 1, ytick_interval = 1,
horizontal_gridlines_on = False,
vertical_gridlines_on = False);

color = 'indigo'

quicklook.add_line_to_chart(chart_skeleton,
x = np.linspace(-1,1,10),
y = np.linspace(-1,1,10),
color_name = color,
color_brightness = 'default',
linewidth = 3,
linestyle = '-',
marker_shape = 'X',
label_for_legend = '',
layer_order = 1)

quicklook.add_line_to_chart(chart_skeleton,
x = np.linspace(-0.5,1,10),
y = np.linspace(-1,1,10),
color_name = color,
color_brightness = 'light',
linewidth = 3,
linestyle = '-',
marker_shape = '.',
label_for_legend = '',
layer_order = 1)

quicklook.add_line_with_error_to_chart(chart_skeleton,
x = np.linspace(-1,1,10),
y_mean = np.linspace(0.5,1,10),
y_error = np.full(10, 0.1),
color_name = 'red',
color_brightness = 'default',
linewidth = 3,
linestyle = '-',
marker_shape = 'o',
label_for_legend = '',
layer_order = 1)

quicklook.add_scatter_to_chart(chart_skeleton,
x = np.linspace(-1,0.5,50) + np.random.rand(50),
y = np.linspace(-1,1,50) + np.random.rand(50),
color_name = 'grape',
color_brightness = 'default',
marker_shape = 'o',
label_for_legend = '',
layer_order = 1)

quicklook.add_scatter_with_error_clouds_to_chart(chart_skeleton,
x = np.linspace(-1,0.5,50) + np.random.rand(50),
y = np.linspace(-1,1,50) + np.random.rand(50),
x_error = np.full(50,0.05),
y_error = np.full(50,0.1),
color_name = 'orange',
color_brightness = 'default',
marker_shape = 'o',
label_for_legend = '',
layer_order = 2)
```
