```python
import quicklook
import numpy as np
```

```python
figure = quicklook.initialize_plot(style = 'default', size = 'default',
title = '',
xlabel = '',
ylabel = '',
x_min_max = (-1,1), y_min_max = (-1,1),
xtick_interval = 1, ytick_interval = 1,
horizontal_gridlines_on = False,
vertical_gridlines_on = False);

color = 'indigo'

quicklook.plot_line(figure,
x = np.linspace(-1,1,10),
y = np.linspace(-1,1,10),
color_name = color,
color_brightness = 'default',
linewidth = 3,
linestyle = '-',
marker_shape = 'X',
label_for_legend = '',
zorder = 1)

quicklook.plot_line(figure,
x = np.linspace(-0.5,1,10),
y = np.linspace(-1,1,10),
color_name = color,
color_brightness = 'light',
linewidth = 3,
linestyle = '-',
marker_shape = '.',
label_for_legend = '',
zorder = 1)

# quicklook.plot_line(figure,
# x = np.linspace(-0,1,10),
# y = np.linspace(-1,1,10),
# color_name = color,
# color_brightness = 'dark',
# linewidth = 3,
# linestyle = '-',
# marker_shape = '',
# label_for_legend = '',
# zorder = 1)

quicklook.plot_line_with_error(figure,
x = np.linspace(-1,1,10),
y_mean = np.linspace(0.5,1,10),
y_error = np.full(10, 0.1),
color_name = 'red',
color_brightness = 'default',
linewidth = 3,
linestyle = '-',
marker_shape = 'o',
label_for_legend = '',
zorder = 1)

# quicklook.plot_vertical_line(figure,
# x = 0,
# color_name = 'teal',
# color_brightness = 'default',
# linewidth = 3,
# linestyle = '-',
# marker_shape = 'None',
# label_for_legend = '',
# zorder = 1)
```
