```python
import quicklook
import numpy as np

# ---- build the chart skeleton; this is chart background, axes, and labels
chart_skeleton = quicklook.build_chart_skeleton(style = 'default', size = 'default',
title = 'Visualize Your Data With quicklook',
xlabel = 'Some Arbitrary Label',
ylabel = 'Another\nArbitrary Label',
x_min_max = (0,1), y_min_max = (0,1),
xtick_interval = 0.5, ytick_interval = 0.25,
horizontal_gridlines_on = False,
vertical_gridlines_on = False);

# ---- add orange dots to the chart
quicklook.add_scatter_to_chart(chart_skeleton,
x = np.random.uniform(0.5,1,20),
y = np.random.uniform(0,0.75,20),
color_name = 'orange',
color_brightness = 'default',
marker_shape = 'o',
opacity = 1,
label_for_legend = 'Orange Dots',
layer_order = 3)

# ---- add blue clouds to the chart
quicklook.add_scatter_with_error_to_chart(chart_skeleton,
x = np.random.uniform(0,0.5,10),
y = np.random.uniform(0.25,1,10),
x_error = np.full(10,0.05),
y_error = np.full(10,0.05),
color_name = 'indigo',
color_brightness = 'default',
marker_shape = 'o',
label_for_legend = 'Blue Clouds',
layer_order = 2)

# ---- add a line to the chart
quicklook.add_line_to_chart(chart_skeleton,
x = np.linspace(0,1,100),
y = 1/(1+np.exp(-8*np.linspace(-1,1,100))),
color_name = 'gray',
color_brightness = 'dark',
linewidth = 3,
linestyle = '-.',
marker_shape = '',
opacity = 1,
label_for_legend = '',
layer_order = 1)
```
